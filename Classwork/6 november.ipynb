```python
def water_after_last_pour(input_str: str) -> str:
    lines = input_str.strip().splitlines()
    n = int(lines[0])

    current_water = 0
    prev_time = 0

    for i in range(1, n + 1):
        t, v = map(int, lines[i].split())

        dt = t - prev_time

        current_water = max(0, current_water - dt)

        current_water += v

        prev_time = t

    return str(current_water)

print(water_after_last_pour("""10
2 1
22 10
26 17
29 2
45 20
47 32
72 12
75 1
81 31
97 7"""))
```
