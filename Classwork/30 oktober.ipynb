Операции со скалярными массивами.
Векторное (линейное пространство) - математическая структура, кот. предсьавляет собой набор элементов (векторов), для кот. определены операции сложения друг с другом, умнож. на число (скаляр), операции подчинены 8 аксиомам:

V — непустое множество элементов произвольной природы, которые называются векторами.

1. x + y = y + x; x, y $\in$ V (коммутативность сложения);
2. x + (y + z)=(x + y) + z для любых x, y, z $\in$ V (ассоциативность сложения);
3. существует такой элемент 0 $\in$ V, что x + 0 = 0 + x = x для любого x $\in$ V (существование нейтрального элемента относительно сложения), называемый нулевым вектором, или просто нулём, пространства V;
4. для любого x $\in$ V существует такой элемент -x $\in$ V, что x +(-x) = 0, называемый вектором, противоположным вектору x;
5. $\alpha$ ( $\beta$ x) = ($\alpha$ $\beta$) x  (ассоциативность умножения на скаляр);
6. 1 * x = x (унитарность: умножение на нейтральный (по умножению) элемент поля F сохраняет вектор).
7. ($\alpha$ + $\beta$ ) x = $\alpha$ x + $\beta$ x (дистрибутивность умножения вектора на скаляр относительно сложения скаляров);
8. $\alpha$ (x + y) = $\alpha$ x + $\alpha$ y (дистрибутивность умножения вектора на скаляр относительно сложения векторов).


Правило треугольника и параллелограмма.

Правило треугольника для сложения векторов: 
оба вектора переносятся параллельно самим себе так, чтобы начало одного из них совпадало с концом другого. Тогда вектор суммы задаётся третьей стороной образовавшегося треугольника, причём его начало совпадает с началом первого вектора, а конец — с концом второго вектора.
Правило параллелограмма для сложения векторов: 
оба вектора переносятся параллельно самим себе так, чтобы их начала совпадали. Тогда вектор суммы задаётся диагональю построенного на них параллелограмма, исходящей из их общего начала. 

$a~\cdot~b = (a_1, a_2, ..., a_n) \cdot (b_1, b_2, ..., b_n) = a_1b_1 + a_2b_2 + ... + a_nb_n = \sum_{=1}^{n} a_ib_i$


Изначальный вектор не изменится:
v1 = np.arange(0, 5)
v1 * 2

Если есть присвоение, то вектор изменится:
v1 = np.arange(0, 5)
v1 = v1 + 2

Поэлементные операции с массивами с совместимыми формами (совпадающая размерность) - по умолчанию операции (сложение, умножение и тд) поэлементно. Пример: матрица (5, 5) может быть умножена на вектор (5, 0).

ИЗМЕНЕНИЕ ФОРМЫ, МАССИВА, СТАККИНГ

изменение формы (размерности массива, матрицы) можно сделать без изменения массива:

B = A.reshape((1, n*m))
- старый массив А не изменился

np.reshape(array, new_shape, order='C')
- array — исходный массив NumPy
- new_shape — кортеж целых чисел, задающий новую форму. Например: (2, 3) или (2, 2, 7)
- order (опционально) — порядок заполнения элементов:
  'C' (по умолчанию) — построчный (C-style),
  'F' — постолбцовый (Fortran-style),
  'A' — сохраняет порядок исходного массива.

Примеры:
1. arr = np.array([1, 2, 3, 4, 5, 6])
   # всего 6 элементов
   reshaped = arr.reshape(2, 3)
   # две строки по три элемента в каждой строке (или 3 столбца), 2*3 = 6 элементов всего
   print(reshaped)

   [[1 2 3]
   [4 5 6]]

2. arr = np.arange(8)
   # всего 8 элементов
   reshaped = arr.reshape(2, 2, 2) 
   # два блока (первое измерение), две строки(второе измерение), все они по два элемента в каждой строке (или 2 столбца) (третье измерение), 2*2*2 = 8 элементов всего
   print(reshaped)

   [[[0 1]
   [2 3]]

   [[4 5]
   [6 7]]]

Если один из размеров указан как -1, NumPy автоматически вычислит его на основе общего числа элементов.
 пример: arr = np.arange(12)
         arr.reshape(3, -1)  # → форма (3, 4)

Для высокой производительности нужно избегать циклического перебора элементов, вместо этого использовать векторизованные алгоритмы.
Первый шаг преобразования скалярного алгоритма в векторизованный - это убедиться, что собственная функция работает с векторными входами.

Векторизация = «Сделай операцию сразу для всех элементов»

Невекторизованная версия = «Возьми первый элемент → обработай → возьми второй → обработай → …»

1. векторизовать существующую функцию:

Theta_vec = np.vectorize(Theta)

2. с самого начала векторизовать функцию:

def Theta(x):
    return 1 * (x >= 0)

Библиотека linalg (from numpy import linalg).

Это «калькулятор» для матричных вычислений:
- решает системы линейных уравнений;
- находит обратные матрицы;
- вычисляет определители;
- ищет собственные значения и векторы;
- выполняет разложения матриц (SVD, QR и др.);
- проверяет свойства матриц (вырожденность, ортогональность).

1. np.linalg.det(A) - определитель матрицы
2. np.linalg.inv(A) - найти обратную матрицу
3. Решить систему уравнений Ax=b
   b = np.array([5, 6])
   x = np.linalg.solve(A, b)
4. Найти собственные значения:
   eigenvals, eigenvecs = np.linalg.eig(A)
   print("Собственные значения:", eigenvals)
