Линейная алгебра (LinAl)

1. Операции со скалярными массивами (векторами-упорядоченный набор чисел, массив)

а = (а1, а2, а3, ..., аn): 
а1 - материал дома,
а2 - соседи,
а3 - близость до метро и т.д.

Сложение, вычитание:
a +/- b =(a1,a2,...,an) + (b1,b2,...,bn) = (a1/-b1,a2/-b2,...,an/-bn)
вектор +/- вектор = вектор

def vector(list1,list2):
    return [a+b for a,b in zip(list1,list2)]

a =[1,2,3,4,5]
b =[6,2,3,4,5]
print(vector(a,b))

Умножение, деление:
1. Поэлементное умножение(произведение):
a*b = (a1,a2,...,an)*(b1,b2,...,bn) = (a1*b1,a2*b2,...,an*bn) 
вектор * вектор = вектор
2. Скалярное произведение(пример - косинусное расстояние, схожесть):
a*b = (a1,a2,...,an)*(b1,b2,...,bn) = a1*b1 + a2*b2 +...+ an*bn
вектор * вектор = скаляр
3. Векторное произведение

4. Деление:
a/b = (a1,a2,...,an)/(b1,b2,...,bn) = (a1/b1,a2/b2,...,an/bn) 
вектор / вектор = вектор

NUMPY

Для векторов, матриц и многомерных наборов данных используется термин МАССИВ.

Функция np.array - создание списков,матриц (большая буква M)

<имя списка,матрицы>.shape или np.shape(<имя списка,матрицы>) - узнать размерность вектора или матрицы
<имя списка,матрицы>.size или np.size(<имя списка,матрицы>)- количество элементов в массиве

Массивы numpy статически типизированы и гомогенны (однородны), тип элемента определяется при создании массива, эффективны с точки зрения использования памяти, благодаря статической типизации быстрая реализация матем. функций(сложение, умножен...) с массивами может быть реализована на скомпилированном языке (Фортран, СИ).

<имя списка,матрицы>.dtype - узнать тип данных

<имя списка,матрицы> = np.array([[1,2],[3,4]], dtype=complex) - для явного указания типа данных внутри массива (dtype=)

Типы данных:
int,
float,
complex,
bool,
object
Можно определить тип данных в битах: например 'int64'

Функции для генерации:

1. <имя списка,матрицы> = np.arange(start, stop, step) - создание массива
2. <имя списка,матрицы> = np.linspace(0, 10, 25) - последовательность данных, равномерно расположенных на числовой прямой в заданном интервале, (start,stop, кол-во значений), в примере: 25 значений на интервале от 0 до 10
3. <имя списка,матрицы> = np.logspace(0, 10, 10, base=np.e) - последовательсть данных, которые равномерно расположены в заданном интервале на числовой прямой, но в лагорифм. масштабе
4. <имя списка,матрицы>, <имя списка,матрицы> = np.mgrid[0:5, 0:5] - специальный тип массива данных в numpy, который создает двухмерный массив 2D с аналогичными значениями, сетка массива, все одного размера и размерности, полезно для создания координат
5. np.random.rand(5,5) - генерация случайных матриц
   np.random.randn(5,5)  - генерация случайных матриц c нормальным распределением
6. создание диагональной матрицы:
   np.diag([1,2,3,4])
   np.diag([1,2,3,4], k = 1) # смещение матрицы направо 
7. Матрица только с нулями - np.zeros((3, 3)) 
   Матрица только с единицами- np.ones((3,3))
