Задача 6. Векторизация  
Напишите функцию f(x), которая возвращает 1, если x >= 0 и 0 иначе.  
Примените эту функцию к массиву np.array([-5, -2, 0, 3, 7]) с помощью векторизации.


```python
def f(x):
    if x >= 0:
        return 1
    else:
        return 0

f_vect = np.vectorize(f)
f_vect(np.array([-5, -2, 0, 3, 7]))

```
