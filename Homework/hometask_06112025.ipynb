Задание 3

4x + 2y + z = 4  
x + 3y = 12  
5y + 4z = -3

```python
A = np.array([
            [4, 2, 1],
            [1, 3, 0],
            [0, 5, 4]
            ]) 

b = np.array([4, 12, -3])

x = np.linalg.solve(A, b)
x
```

Задание 9  
Напишите функцию, которая на вход принимает 2 вектора и возвращает тот из них, что больше. Функцию необходимо векторизовать.

```python

def max_function(a, b):
    if np.sum(a) > np.sum(b):
        print(f'Вектор a = {a} больше вектора b = {b}')
    else:
        print(f'Вектор b = {b} больше вектора a = {a}')

max_function_vec = np.vectorize(max_function)

max_function([1, 2, 3, 50], [4, 5, 6])
    
```
